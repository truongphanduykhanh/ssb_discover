```python
import json

import pandas as pd
import requests
from pyjstat import pyjstat

import eda
```

```python
zone_room_space_url = "https://data.ssb.no/api/v0/en/table/09897/"
zone_room_space_query = {
  "query": [],
  "response": {
    "format": "json-stat2"
  }
}

zone_room_url = "https://data.ssb.no/api/v0/en/table/09895/"
zone_room_query = {
  "query": [
    {
      "code": "Soner2",
      "selection": {
        "filter": "item",
        "values": [
          "00",
          "01",
          "02",
          "03",
          "04",
          "05",
          "20",
          "21",
          "22"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}

tenancy_url = "https://data.ssb.no/api/v0/en/table/06230/"
tenancy_query = {
  "query": [],
  "response": {
    "format": "json-stat2"
  }
}

landlords_url = "https://data.ssb.no/api/v0/en/table/09896/"
landlords_query = {
  "query": [],
  "response": {
    "format": "json-stat2"
  }
}
```

```python
def query_to_df(url: str, query: dict):
	res = requests.post(url, json=query)
	ds = pyjstat.Dataset.read(res.text)
	df = ds.write('dataframe')
	return df
```

```python
zone_room_space = query_to_df(zone_room_space_url, zone_room_space_query)
zone_room = query_to_df(zone_room_url, zone_room_query)
tenancy = query_to_df(tenancy_url, tenancy_query)
landlords = query_to_df(landlords_url, landlords_query)
```
